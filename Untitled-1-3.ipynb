```python
plt.figure(figsize = (10,5))
sns.set(font_scale=1.5)
sns.regplot(
    x= 'week',
    y= 'covid_deaths',
    data = df,
    fit_reg = True
)
plt.show()
```

```python
sns.boxplot(data = df.loc[:,['covid_deaths']])

plt.show()
```

```python
plt.hist(df.loc[(df['covid_deaths'] >= 0)&(df['covid_deaths'] <= 2000),
                'covid_deaths'
               ]
        )

plt.rcParams['figure.figsize'] = (50,50)

plt.rcParams['font.size'] = 50

plt.xlabel('Número de Muertes',
           fontsize= 60
          )

plt.ylabel('Frecuencia', 
           fontsize= 60
          )

plt.title('HISTOGRAMA DE MUERTES POR COVID', 
          fontsize=100
         )
```
