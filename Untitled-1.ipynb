```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv ('all_weekly_excess_deaths.csv', decimal= '.')
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
sns.lmplot(
    
    y= 'week',
    x= 'covid_deaths',
    data = df,
    fit_reg = True,
    hue = 'country'
)
```

```python
sns.boxplot(data = df.loc[:,['covid_deaths']])

plt.show()
```

```python
plt.hist(df.loc[(df['covid_deaths'] >= 0)&(df['covid_deaths'] <= 2000), 'covid_deaths'])
plt.rcParams['figure.figsize'] = (50,50)
plt.rcParams['font.size'] = 50
plt.xlabel('Número de Muertes', fontsize= 60)
plt.ylabel('Frecuencia', fontsize= 60)
plt.title('HISTOGRAMA DE MUERTES POR COVID', fontsize=100);
```
